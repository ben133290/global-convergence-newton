```python
training_losses = []
gradient_norms = []

for run in range(number_of_runs_per_method):
    lr = perform_train_run("covtype", "ce", "m22", epochs=10)
    training_losses.append(lr.losses)
    gradient_norms.append(lr.grad_norm)
```

```python
def average_runs_per_epoch(runs):
    runs = np.array(runs)  # shape: (num_runs, epochs, classes)
    return np.mean(runs, axis=0)

def plot_loss_and_grad_norm(training_losses, gradient_norms, method_name="M22"):
    # Average over runs per epoch
    avg_loss = average_runs_per_epoch(training_losses)
    avg_grad = average_runs_per_epoch(gradient_norms)
    
    epochs = np.arange(avg_loss.shape[1])

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    for class_idx in range(avg_loss.shape[0]):
        ax1.plot(epochs, avg_loss[class_idx, :], label=f"Class {class_idx}")
    ax1.set_title(f"Training Loss (Averaged) - {method_name}")
    ax1.set_xlabel("Epoch")
    ax1.set_ylabel("Loss")
    ax1.legend()
    ax1.grid(True)

    for class_idx in range(avg_grad.shape[0]):
        ax2.plot(epochs, avg_grad[class_idx, :], label=f"Class {class_idx}")
    ax2.set_title(f"Gradient Norm (Averaged) - {method_name}")
    ax2.set_xlabel("Epoch")
    ax2.set_ylabel("Gradient Norm")
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

```

```python
print(average_runs_per_epoch(training_losses))
print(np.arange(average_runs_per_epoch(training_losses).shape[1]))
```

```python
plot_loss_and_grad_norm(training_losses, gradient_norms)
```
