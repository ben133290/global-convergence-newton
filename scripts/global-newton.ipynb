```python
from main import perform_train_run
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_runs_per_method = 5
epochs = 10
methods = ['gd', 'newton', 'm22', 'cubic', 'ada']
```

#### Design a run:
- Choose a method from methods
- Choose a loss function 'ce' or 'ncce'
- Choose a dataset from 'a9a', 'covtype', 'ijcnn1', 'mnist'
- The rest of the parameters should be self-explanitory
- update the parameters of perform_train_run(dataset, loss, method, epochs=n)

#### Evaluation:
Note that the test accuracy is outputed at the end of the training run, so you might need to scroll down to the end of the output to see it.

```python
all_training_losses = []
all_gradient_norms = []
all_training_accuracies = []

for i, method in enumerate(["gd", "grd", "adan", "newton"]):
    training_losses = []
    gradient_norms = []
    training_accuracies = []
    for run in range(number_of_runs_per_method):
        lr, accuracy = perform_train_run("ijcnn1", "ce", method, epochs=epochs)
        training_losses.append(lr.losses)
        gradient_norms.append(lr.grad_norm)
        training_accuracies.append(lr.train_accuracies)
    all_training_losses.append(training_losses)
    all_gradient_norms.append(gradient_norms)
    all_training_accuracies.append(training_accuracies)
```

```python
training_losses_gd = []
gradient_norms_gd = []
training_accuracies_gd = []

for run in range(number_of_runs_per_method):
    lr2, accuracy2 = perform_train_run("mnist", "ce", "gd", epochs=epochs)
    training_losses_gd.append(lr2.losses)
    gradient_norms_gd.append(lr2.grad_norm)
    training_accuracies_gd.append(lr2.train_accuracies)
```

```python
def average_runs_per_epoch(runs):
    runs = np.array(runs)  # shape: (num_runs, epochs, classes)
    return np.mean(runs, axis=0)

def plot_stuff(training_losses, gradient_norms, train_acc, method_name="M22", data_set="IJCNN1"):
    # Average over runs per epoch
    avg_loss = average_runs_per_epoch(training_losses)
    avg_grad = average_runs_per_epoch(gradient_norms)
    avg_acc = average_runs_per_epoch(train_acc)

    ax1.plot(avg_loss, label=f"{method_name}")

    ax2.plot(avg_grad, label=f"{method_name}")

    ax3.plot(avg_acc, label=f"{method_name}")
```

```python
print(average_runs_per_epoch(training_losses))
print(average_runs_per_epoch(training_accuracies))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 5))
data_set = "MNIST"
loss_type = "CE"

plot_stuff(all_training_losses[1], all_gradient_norms[1], all_training_accuracies[1], method_name="Global Regularized Newton")
plot_stuff(all_training_losses[0], all_gradient_norms[0], all_training_accuracies[0], method_name="Gradient Descent")
plot_stuff(all_training_losses[2], all_gradient_norms[2], all_training_accuracies[2], method_name="Adaptive Newton")
plot_stuff(all_training_losses[3], all_gradient_norms[3], all_training_accuracies[3], method_name="Classic Newton")

ax1.set_title(f"Loss (Averaged) - {data_set}, {loss_type}")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.legend()
ax1.grid(True)

ax2.set_title(f"Gradient Norm (Averaged) - {data_set}, {loss_type}")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Gradient Norm")
ax2.legend()
ax2.grid(True)

ax3.set_title(f"Accuracy (Averaged) - {data_set}, {loss_type}")
ax3.set_xlabel("Epoch")
ax3.set_ylabel("Accuracy")
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```
